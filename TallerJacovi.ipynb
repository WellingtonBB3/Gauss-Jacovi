**Ejercicios con Convergencia**

```python
import numpy as np  
import matplotlib.pyplot as plt

matriz_A = np.array([[3, 2], [2, 10]], dtype=float)
vector_b = np.array([11, 16], dtype=float)
vector_inicial = np.array([[1], [1]])
tolerancia = 1e-5
maximo_iteraciones = 100

solucion_jacobi, trayectoria_jacobi = gauss_jacobi(
    A=matriz_A, b=vector_b, x0=vector_inicial, tol=tolerancia, max_iter=maximo_iteraciones
)
trayectoria_jacobi = np.squeeze(np.array(trayectoria_jacobi))
valores_x1_jacobi = trayectoria_jacobi[:, 0]
valores_x2_jacobi = trayectoria_jacobi[:, 1]

solucion_seidel, trayectoria_seidel = gauss_seidel(
    A=matriz_A, b=vector_b, x0=vector_inicial, tol=tolerancia, max_iter=maximo_iteraciones
)
trayectoria_seidel = np.squeeze(np.array(trayectoria_seidel))
valores_x1_seidel = trayectoria_seidel[:, 0]
valores_x2_seidel = trayectoria_seidel[:, 1]

plt.plot(
    valores_x1_jacobi, 
    valores_x2_jacobi, 
    marker='o', 
    color='blue', 
    label="Trayectoria Gauss-Jacobi", 
    linestyle='-'
)
plt.plot(
    valores_x1_seidel, 
    valores_x2_seidel, 
    marker='^', 
    color='red', 
    label="Trayectoria Gauss-Seidel", 
    linestyle='-.'
)

plt.title("Gauss-Jacobi vs Gauss-Seidel", color='purple')
plt.xlabel("$x_1$", color='darkgreen')
plt.ylabel("$x_2$", color='darkgreen')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt

matriz_A = np.array([[3, 2], [2, 10]], dtype=float)
vector_b = np.array([11, 16], dtype=float)
vector_inicial = np.array([[5], [-2]])
tolerancia = 1e-5
maximo_iteraciones = 100

solucion_jacobi, trayectoria_jacobi = gauss_jacobi(
    A=matriz_A, b=vector_b, x0=vector_inicial, tol=tolerancia, max_iter=maximo_iteraciones
)
trayectoria_jacobi = np.squeeze(np.array(trayectoria_jacobi))
valores_x1_jacobi = trayectoria_jacobi[:, 0]
valores_x2_jacobi = trayectoria_jacobi[:, 1]

solucion_seidel, trayectoria_seidel = gauss_seidel(
    A=matriz_A, b=vector_b, x0=vector_inicial, tol=tolerancia, max_iter=maximo_iteraciones
)
trayectoria_seidel = np.squeeze(np.array(trayectoria_seidel))
valores_x1_seidel = trayectoria_seidel[:, 0]
valores_x2_seidel = trayectoria_seidel[:, 1]

plt.plot(
    valores_x1_jacobi, 
    valores_x2_jacobi, 
    marker='o', 
    color='cyan', 
    label="Trayectoria Gauss-Jacobi", 
    linestyle='-'
)
plt.plot(
    valores_x1_seidel, 
    valores_x2_seidel, 
    marker='^', 
    color='orange', 
    label="Trayectoria Gauss-Seidel", 
    linestyle='-.'
)

plt.title("Gauss-Jacobi vs Gauss-Seidel", color='blue')
plt.xlabel("$x_1$", color='darkred')
plt.ylabel("$x_2$", color='darkred')
plt.grid()
plt.legend()
plt.show()
```
